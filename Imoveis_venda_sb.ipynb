#Função para extrair os dados, e retornar uma lista python

```python
def tabela_dados(html: str) -> list:
  html = html
  soup = BeautifulSoup(html, "html.parser")
  data_frame = []
  for imovel in soup.find_all("div", attrs={"class": "imovel_resultado sprite-bg_search_large"}):
    imoveis = []
    dados = imovel.text
    try:
      separar_dados = [i for i in dados.split("\n") if i] # separando as informações pelo "\n"
      formatar_localizacao = re.findall(r'.*\-(.*).?\,.[0-9]{2,4}.\-(.*?)\-.(.*?)\-(.*?)\-(.*?)|.*\-(.*?)\-.(.*?)\-(.*?)\-(.*?)', separar_dados[3])
      formatar_referencia = re.findall(r'\-.RS.\-(.*)$', separar_dados[3])
      comodos = [i for i in formatar_referencia[0].split(',') if i]
      formatar_valor = re.findall(r'.*?R\$.([0-9]{1,3}\.[0-9]{3})|.*$', separar_dados[2])
    except Exception as e:
      print(e)
    finally:
      # adicionando os dados tradados a uma lista
      imoveis.append(separar_dados[1]) # coluna tipo_imovel
      imoveis.append(formatar_localizacao[0][0]) # coluna endereco
      imoveis.append(formatar_localizacao[0][1]) # coluna bairro
      imoveis.append(formatar_localizacao[0][2]) # coluna cidade
      imoveis.append(formatar_localizacao[0][3]) # coluna estado
      imoveis.append(formatar_referencia[0]) # coluna comodos 
      imoveis.append(formatar_valor[0]) # coluna valor
      # lista final que será um dataframe pandas
      data_frame.append(imoveis)
  return data_frame
```

# Fazendo uma junção (join), nas páginas extraida.

```python
lista_dados1 = tabela_dados(pagina_html1)
lista_dados2 = tabela_dados(pagina_html2)
lista_dados3 = tabela_dados(pagina_html3)
lista_dados4 = tabela_dados(pagina_html4)
lista_dados5 = tabela_dados(pagina_html5)
join_paginas_html = []

for lista in range(len(lista_dados1)):
    join_paginas_html.append(lista_dados1[0])
    lista_dados1.pop(0)
    join_paginas_html.append(lista_dados2[0])
    lista_dados2.pop(0)
    join_paginas_html.append(lista_dados3[0])
    lista_dados3.pop(0)
    join_paginas_html.append(lista_dados4[0])
    lista_dados4.pop(0)
    if lista_dados5 != []:
      join_paginas_html.append(lista_dados5[0])
      lista_dados5.pop(0)
    else:
      continue
```

#Gerando um DataFrame com a lista de dados dos imoveis

```python
df = pd.DataFrame(join_paginas_html, columns=('tipo_imovel', 'endereco', 'bairro', 'cidade', 'estado', 'comodos', 'valor'))
```
